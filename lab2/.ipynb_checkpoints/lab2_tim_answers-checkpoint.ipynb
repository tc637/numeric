```python
theFuncs={'euler':euler,'leapfrog':leapfrog,'runge':runge}# 'midpoint':midpoint}

if __name__=="__main__":
    Ta= 20
    To= 30
    tend = 30
    theLambda= -0.8
    npts=[60, 120, 240, 480, 960] # number of points, originally 30
    local_error = []
    global_error = []
    points = []
    #
    #find the method in the theFuncs dictionary and call it
    #
    
    # plot 3 figures, each with two subplots; the first with the exact and approx solutions for 
    # each timestep for one numerical scheme, the other with the errors
    for funChoice in theFuncs: # loop through each function
        plt.close('all')
        fig1,(ax1,ax2)=plt.subplots(2,1)
        for npt in npts: # loop through each total number of points
            approxTime,approxTemp=theFuncs[funChoice](npt,tend,To,Ta,theLambda)
            ax1.plot(approxTime,approxTemp,label="{:.2}".format(tend/npt))
            exactTime=np.empty([npt,],np.float)
            exactTemp=np.empty_like(exactTime)
            
            for i in np.arange(0,npt):
                exactTime[i] = tend*i/npt
                exactTemp[i] = Ta + (To-Ta)*np.exp(theLambda*exactTime[i])
        
            difference = np.abs(exactTemp - approxTemp)
            ax2.plot(exactTime,difference,label="{:.2}".format(tend/npt))
            
            if npt == npts[0]:
                max_difference = max(difference) # grab the max difference to set upper ylim
            
            if funChoice == 'runge':
                local_error.append(difference[1])
                global_error.append(difference[-1])
                points.append(npt)
            
        ax1.plot(exactTime,exactTemp,'m+',label='exact')
        ax1.set_xlim(0,tend)
        outdict=dict(deltat=tend/npt,func=funChoice)
        title="exact and approx using {func} ".format_map(outdict)    
        ax1.set(title=title)
        ax1.legend(loc='best')
        ax2.legend(loc='best')
        ax2.set(ylabel='exact - approx')
        ax2.set_ylim(0,max_difference)
        plt.show()
```

```python
# a.) Decreasing the step-size and increasing the order of the scheme do not always improve the solution. As shown above, with 
#     100000 points, the error is smaller for both leapfrog and runge compared to 1000000 points. At those npts, Euler is also
#     more accurate than leapfrog.
# b.) Global truncation error at a point can be read straight from the error curve at that point, i.e. the value of the error
#     at t_n. The local error can be determined by initializing an approximation at a point t_i, calculating the error after 
#     one step.

# for b.)
print("Local error at t_1 with npt = {} for runge is {}".format(points[0], local_error[0]))
print("Global error for runge with tend = {} and npt = {} is {}".format(tend, points[0], global_error[0]))

# c.) Read error for several different timesteps at a point to get the order. For the global error, do the same but 
#     at the endpoint.

# Local
plt.close('all')

steps = [tend/x for x in points]
fig1,ax1 = plt.subplots(1,1)
ax1.plot(steps,local_error,'.m')
ax1.set(title='local error for runge')
ax1.set(xlabel='timestep')
ax1.set(ylabel='absolute error')
ax1.set_ylim(0,max(local_error)+0.5*(max(local_error)))
ax1.set_xlim(0,0.6)

# Global

fig2,ax2 = plt.subplots(1,1)
ax2.plot(steps,global_error,'.c')
ax2.set(title='global error for runge')
ax2.set(xlabel='timestep')
ax2.set(ylabel='absolute error')
ax2.set_ylim(0,max(global_error)+0.5*(max(global_error)))
ax2.set_xlim(0,0.6)
plt.show()

# Global error is of one order lower than local, i.e. less accurate because it is the sum of all local errors at each timestep
# before t_n. For runge, local error is fourth order, while global error is third order.
```

```python
# Problem Stability

"""
  this script show how to plot the heat conduction equation
"""

import matplotlib.pyplot as plt
from numlabs.lab2.lab2_functions import euler,beuler,leapfrog
import numpy as np


theFuncs={'euler':euler,'beuler':beuler,'leapfrog':leapfrog}

if __name__=="__main__":
    tend=30.
    Ta=20.
    To=30.
    theLambda=-8.
    funChoice='euler'
    npts = [tend*4,tend*4.5,tend*5,tend*6] # multiples of Lambda
    for funChoice in theFuncs: # loop through all functions
        plt.close('all')
        fig,ax = plt.subplots(1,1)
        for npt in npts: # loop through each total number of points
            approxTime,approxTemp=theFuncs[funChoice](npt,tend,To,Ta,theLambda)
            exactTime=np.empty([npt,],float)
            exactTemp=np.empty_like(exactTime)
            for i in np.arange(0,npt):
                exactTime[i] = tend*i/npt
                exactTemp[i] = Ta + (To-Ta)*np.exp(theLambda*exactTime[i])
            if npt == npts[0]:
                plt.plot(exactTime,exactTemp,'m+',label='exact')
            plt.hold(True)
            plt.plot(approxTime,approxTemp,label="{:.2} (tend/{:.2})".format(tend/npt, (npt/tend)))
            
        ax.legend(loc='best')   
        ax.set_xlim([0,30])
        ax.set_ylim([15,30])
        ax.set_title('stability of {}'.format(funChoice))
        plt.show()

print("""The Euler scheme quickly converges to the exact solution for small values of delta_t. However, 
at tend/4 (i.e. delta_t = -2/lambda = -2/-8), the Euler scheme results in huge oscillations, which shows 
that instability occurs when the instability condition is reached.\n""")

print("""In contrast, the leapfrog method is always unstable (for all tested values of the timestep). However, the
backward Euler method is always stable for all timesteps.""")

```
